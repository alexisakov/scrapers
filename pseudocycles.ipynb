```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
x = np.random.normal(loc=0.0, scale=3.0, size=400)
xma = moving_average(x, 20)
```

```python
plt.plot(list(range(len(x))), x)
plt.plot(np.array(list(range(len(xma))))+10, xma)

plt.show()
```
